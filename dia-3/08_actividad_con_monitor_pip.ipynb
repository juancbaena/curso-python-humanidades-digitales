# PARTE 1 — Instalar y verificar (monitoría del Día 3)

## Regla de oro (memorízala)

En Windows, **NO** escribas `pip install ...` a secas. Escribe siempre:

```
py -m pip install NOMBRE_DEL_PAQUETE
```

¿Por qué? Porque muchas computadoras tienen **más de un Python instalado**. Si escribes `pip` solo,
no sabes a *cuál* Python le estás instalando el paquete, y luego tu programa "no lo encuentra".
La fórmula `py -m pip` le dice a Windows: *"usa el lanzador oficial `py` y, con ESE Python, ejecuta pip"*.
Así el paquete queda donde sí lo vas a usar.

> En **macOS/Linux** la idea es la misma, pero se escribe `python3 -m pip install NOMBRE`.


## Paso 1 — Abre la terminal correcta

1. Abre **VS Code**.
2. Menú **Terminal → New Terminal** (o `Ctrl + ñ`). Se abrirá una terminal abajo.
3. Comprueba que Python responde. Escribe y presiona Enter:

```
py --version
```

Deberías ver algo como `Python 3.12.x`. Si en su lugar dice *"py no se reconoce..."*, ve a la
sección **Errores comunes → Error A**.


## Paso 2 — Mira CUÁNTOS Python tienes (esto evita el 80% de los problemas)

Escribe:

```
py -0
```

Esto **lista todas las versiones de Python** instaladas con el lanzador. Si ves más de una, no te
preocupes: justamente por eso usamos `py -m pip`, que siempre apunta a la primera (la predeterminada).

Para ver todos los Python que hay en el PATH del sistema:

```
where python
```

> 🧭 **Para docentes.** "Tengo varios Python" es la causa #1 de que algo "se instale pero no
> funcione". Mostrar `py -0` al inicio le ahorra horas de frustración a tus estudiantes.


## Paso 3 — Actualiza `pip` (recomendado)

```
py -m pip install --upgrade pip
```

Un `pip` viejo a veces falla al instalar. Actualizarlo primero es un buen seguro.


## Paso 4 — Instala `pandas` y `matplotlib`

```
py -m pip install pandas matplotlib
```

Verás muchas líneas y, al final, algo como `Successfully installed ...`. Eso es **éxito**.
Puede tardar un par de minutos: es normal.


## Paso 5 — Verifica que SÍ quedaron instalados

Esta es la prueba definitiva. En la **misma terminal**, escribe:

```
py -c "import pandas, matplotlib; print('OK', pandas.__version__, matplotlib.__version__)"
```

Si imprime `OK` con dos números de versión, **lo lograste**. 🎉

> Si en cambio aparece `ModuleNotFoundError`, ve a **Errores comunes → Error C**.


## Paso 6 — Que VS Code use el Python correcto

Para que tu código dentro de VS Code "vea" lo que instalaste:

1. Presiona `Ctrl + Shift + P`.
2. Escribe **Python: Select Interpreter** y selecciónalo.
3. Elige el Python que aparece como **recomendado** (el mismo de `py`).

> Si no aparece la opción, instala la **extensión Python** (de Microsoft) desde el panel de Extensiones.


### ✅ Prueba final dentro de un cuaderno

La siguiente celda funciona en Colab (ya tiene todo) **y** en tu VS Code local si seguiste los pasos.
Ejecútala: si ves las versiones, todo está listo.


````python
import pandas as pd
import matplotlib
print("pandas:", pd.__version__)
print("matplotlib:", matplotlib.__version__)
print("¡Todo listo para trabajar con datos!")
````

## 🛟 Errores comunes en Windows (y cómo salir de ellos)

### Error A — *"py no se reconoce como un comando..."* o *"python no se reconoce..."*
El lanzador no está disponible. Soluciones:
- Reinstala Python desde <https://www.python.org/downloads/> y, **muy importante**, marca la casilla
  **"Add python.exe to PATH"** en la primera pantalla del instalador.
- Cierra y vuelve a abrir VS Code después de instalar (para que tome el nuevo PATH).
- Si `py` no existe pero `python` sí, usa `python -m pip install ...` en lugar de `py -m pip ...`.

### Error B — *"pip no se reconoce como un comando"*
Estás usando `pip` a secas. **Nunca** lo hagas así. Usa siempre:
```
py -m pip install NOMBRE
```

### Error C — Instala bien, pero al ejecutar dice `ModuleNotFoundError: No module named 'pandas'`
Clásico de **varios Python**: instalaste en uno y estás ejecutando con otro. Soluciones:
- Instala con `py -m pip install pandas` y ejecuta con `py mi_programa.py` (el **mismo** `py`).
- En VS Code, repite el **Paso 6** (Select Interpreter) y elige el Python correcto.

### Error D — *"Defaulting to user installation because normal site-packages is not writeable"*
Es solo un aviso, no un error. Si la instalación termina en `Successfully installed`, está bien.
Si falla por permisos, **no uses** comandos de administrador; mejor usa un **entorno virtual** (Parte 2).

> 🧭 **Para docentes.** Ten a la mano esta lista de errores; son los que tus estudiantes encontrarán.
> Anticiparlos convierte un momento de pánico ("no me funciona nada") en un paso esperado del proceso.

# PARTE 2 — Entornos virtuales (monitoría del Día 4)

## ¿Qué es un entorno virtual y por qué a veces es obligatorio?

Un **entorno virtual** (venv) es una "cajita" aislada con su propio Python y sus propios paquetes, solo
para **un** proyecto. Sirve para que las librerías de un trabajo no choquen con las de otro.

En algunas computadoras, `pip` **se niega a instalar** fuera de un entorno virtual y muestra un error como
`error: externally-managed-environment`. No es un fallo tuyo: el sistema te está pidiendo usar un venv.
La solución es justamente crear uno.

## Receta a la fija (Windows)

Dentro de la carpeta de tu proyecto, en la terminal de VS Code:

**1. Crear el entorno** (solo una vez por proyecto):
```
py -m venv .venv
```

**2. Activarlo:**
- En **PowerShell** (la terminal azul por defecto de VS Code):
  ```
  .\.venv\Scripts\Activate.ps1
  ```
- En **Símbolo del sistema (cmd)**:
  ```
  .venv\Scripts\activate.bat
  ```

Cuando esté activo, verás `(.venv)` al inicio de la línea de la terminal.

**3. Instalar ya dentro del entorno** (aquí sí puedes usar `pip` directo, porque apunta al venv):
```
pip install pandas matplotlib
```

**4. Verificar:**
```
python -c "import pandas, matplotlib; print('OK en el entorno')"
```

**5. Decirle a VS Code que use el entorno:** `Ctrl + Shift + P` → **Python: Select Interpreter** →
elige el que diga **`.venv`**.

> Para **salir** del entorno cuando termines: escribe `deactivate`.


## 🛟 Errores comunes con entornos virtuales

### Error E — *"No se puede cargar el archivo Activate.ps1 porque la ejecución de scripts está deshabilitada en este sistema"*
PowerShell bloquea los scripts por seguridad. Dos salidas:
- **La fácil:** abre una terminal **cmd** en vez de PowerShell (menú desplegable `+` de la terminal en
  VS Code → *Command Prompt*) y activa con `.venv\Scripts\activate.bat`.
- **La definitiva:** en PowerShell, ejecuta una sola vez:
  ```
  Set-ExecutionPolicy -Scope CurrentUser -ExecutionPolicy RemoteSigned
  ```
  Acepta con `S`/`Y` y vuelve a intentar la activación.

### Error F — Activé el entorno pero `import pandas` sigue fallando en VS Code
VS Code todavía apunta al Python de afuera. Repite el **Paso 5** de la receta (Select Interpreter) y
elige el del `.venv`. Si tienes un cuaderno abierto, revisa también el **kernel** (arriba a la derecha).

### Error G — Creé varios `.venv` y me perdí
No pasa nada. Borra la carpeta `.venv` del proyecto y vuelve a crearla con `py -m venv .venv`.
Un entorno virtual es **desechable**: se puede recrear en segundos.
