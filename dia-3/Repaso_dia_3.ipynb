## Actividad 4 — Capturar un archivo inexistente

````python
# Tu solución:
````

<details>


## Actividad 6 — Procesar un archivo del repositorio

Descarga `fechas_historicas.csv` (te damos el código), recórrelo saltando el encabezado e imprime cada evento con su año en el formato `anio -> evento`.

````python
import urllib.request

url = "https://raw.githubusercontent.com/calderonf/curso-python-humanidades-digitales/main/datos/fechas_historicas.csv"
texto = urllib.request.urlopen(url).read().decode("utf-8")
lineas = texto.splitlines()

# Tu solución (usa lineas[1:] para saltar el encabezado):
````

<details>
<summary>👀 Ver solución</summary>

```python
for linea in lineas[1:]:
    partes = linea.split(",")
    evento = partes[0]
    anio = int(partes[1])
    print(f"{anio} -> {evento}")
```
</details>

## Actividad 12 — Gráfico de barras

Haz un gráfico de barras con el número de obras por género (usa `value_counts()` y matplotlib), con título y etiquetas en español.

````python
# Tu solución:
````

<details>
<summary>👀 Ver solución</summary>

```python
import matplotlib.pyplot as plt
conteo = df["genero"].value_counts()
plt.figure(figsize=(8, 4))
plt.bar(conteo.index, conteo.values, color="steelblue")
plt.title("Obras por genero")
plt.xlabel("Genero")
plt.ylabel("Cantidad")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```
</details>
