## ¿Qué es el pensamiento computacional?

Antes que el código está la **forma de pensar** un problema. El pensamiento computacional
tiene cuatro pilares, y los usamos todo el tiempo en humanidades aunque no lo notemos:

| Pilar | En palabras simples | Ejemplo humanístico |
|---|---|---|
| **Descomposición** | Partir un problema grande en pasos pequeños | Armar una ficha bibliográfica: primero el autor, luego el título, luego el año… |
| **Reconocimiento de patrones** | Detectar lo que se repite | Todas las fichas siguen la misma estructura |
| **Abstracción** | Quedarse con lo esencial e ignorar el detalle | Una "referencia" es autor + título + año, sin importar el libro |
| **Algoritmo** | Una secuencia clara de pasos para lograr algo | La receta para construir cualquier ficha |

La celda siguiente muestra la **descomposición** de una tarea cotidiana, escrita como pasos.
No te preocupes por la sintaxis todavía; solo lee:


```python
# Descomponer "armar una referencia bibliografica" en pasos:
print("Paso 1: identificar el autor")
print("Paso 2: identificar el titulo")
print("Paso 3: identificar el año")
print("Paso 4: unir todo en un formato claro")
```

> 🧭 **Nota para docentes — una idea que atraviesa todo el curso**
>
> Cuando ustedes lleven estos temas a sus estudiantes, serán quienes **delimiten los talleres**.
> Una parte esencial de enseñar a programar es **calibrar la dificultad**: proponer retos
> **alcanzables y graduados**, y evitar ejercicios extremadamente difíciles que, al fallar,
> solo consigan **desanimar** a quien apenas empieza.
>
> La meta no es que el ejercicio sea impresionante, sino que el estudiante **termine con una
> pequeña victoria** que lo anime a seguir. A lo largo de la semana volveremos sobre cómo
> dosificar el reto: empezar muy abajo, subir de a poco y poner un "techo" sensato.
