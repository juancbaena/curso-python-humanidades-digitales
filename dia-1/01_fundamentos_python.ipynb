## 1. Mostrar mensajes: `print()` y comentarios

`print()` muestra en pantalla lo que pongas entre paréntesis.
Una línea que empieza con `#` es un **comentario**: Python la ignora, sirve para anotar.


````python
# Esto es un comentario: no se ejecuta.
print("Hola, humanidades digitales")
print("Hoy aprendo a programar")
````

## 2. Variables: guardar información con un nombre

Una **variable** es una etiqueta que guarda un valor para reutilizarlo. Se asigna con `=`.


````python
autor = "Miguel de Cervantes"
titulo = "Don Quijote de la Mancha"
anio = 1605

print(autor)
print(titulo)
print(anio)
````

## 3. Tipos de datos

Cada valor tiene un **tipo**. Los más básicos:

- `str` → texto (cadena), va entre comillas: `"Cervantes"`
- `int` → número entero: `1605`
- `float` → número con decimales: `3.14`
- `bool` → valor lógico: `True` (verdadero) o `False` (falso)

La función `type()` nos dice el tipo de un valor.


````python
print(type(autor))   # texto -> str
print(type(anio))    # entero -> int
print(type(3.14))    # decimal -> float
print(type(True))    # logico -> bool
````

## 4. Operaciones con números

Python funciona como una calculadora: `+`, `-`, `*`, `/`, y `**` (potencia).


````python
anio_actual = 2026
antiguedad = anio_actual - anio
print("Antiguedad de la obra (años):")
print(antiguedad)
````

## 5. Trabajar con texto: unir cadenas y *f-strings*

Podemos **unir** texto con `+`, pero la forma más cómoda es la **f-string**:
se escribe una `f` antes de las comillas y se ponen las variables entre llaves `{ }`.


````python
# Union simple con +
print(autor + " escribió " + titulo)

# Forma recomendada: f-string
referencia = f"{autor} escribió en  ({anio}) la obra {titulo}."
print(referencia)
````

## 6. Pedir datos: `input()`

`input()` detiene el programa y espera a que la persona escriba algo.
**Importante:** lo que devuelve `input()` siempre es **texto (`str`)**. Si necesitamos un número,
lo convertimos con `int()` o `float()`.


````python
titulo_usuario = input("Titulo del libro: ")
autor_usuario = input("Autor: ")
anio_texto = input("Anio de publicacion: ")

anio_numero = int(anio_texto)   # convertimos el texto a numero entero

ficha = f"{autor_usuario} ({anio_numero}). {titulo_usuario}."
print("\nReferencia generada:")
print(ficha)
````

> 💡 **Error muy común (y muy útil para enseñar):** si intentas restar o sumar con el
> resultado de `input()` sin convertirlo, Python se queja porque está mezclando texto y números.
> Ese error es una gran oportunidad para explicar la diferencia entre `str` e `int`.
