## 1. Una cadena es una secuencia de caracteres

Cada carácter de una cadena tiene una **posición** (un *índice*). En Python se empieza a contar
**desde 0**: el primer carácter es la posición `0`, el segundo la `1`, y así sucesivamente.


````python
palabra = "Quijote"

print(palabra[0])   # primer caracter
print(palabra[1])   # segundo caracter
print(palabra[-1])  # el ultimo (los indices negativos cuentan desde el final)
````

## 2. Tomar un pedazo: *slicing*

Con `[inicio:fin]` obtienes un fragmento. Incluye la posición `inicio` y **llega hasta**
`fin` **sin incluirla**.


````python
titulo = "Don Quijote de la Mancha"

print(titulo[0:3])    # "Don"
print(titulo[4:11])   # "Quijote"
print(titulo[-6:])    # "Mancha" (desde la posicion -6 hasta el final)
````

> 💡 Que el conteo empiece en 0 es la **primera sorpresa** clásica para quien aprende.
> Vale la pena detenerse aquí: es un punto donde tus estudiantes se confundirán, y anticiparlo
> evita frustración.


## 3. Métodos de cadena: transformar texto

Un **método** es una acción que un valor "sabe hacer". Se escribe con un punto después de la
variable: `variable.metodo()`. Los métodos de cadena **no cambian** la original; **devuelven una
nueva**.


````python
frase = "  El Quijote es una NOVELA  "

print(frase.upper())     # TODO EN MAYUSCULAS
print(frase.lower())     # todo en minusculas
print(frase.strip())     # quita los espacios sobrantes de los extremos
print(frase.replace("NOVELA", "obra"))  # reemplaza un texto por otro
````

También podemos **contar** cuántas veces aparece algo con `.count()`:


````python
texto = "el mar, el cielo y el viento"
print(texto.count("el"))   # cuantas veces aparece "el"
````

## 4. Medir y separar: `len()` y `.split()`

- `len(texto)` devuelve la **cantidad de caracteres**.
- `texto.split()` parte el texto en **palabras** (usando los espacios) y devuelve una lista.
  Si combinamos `len()` con `split()`, contamos **palabras** sin necesidad de bucles.


````python
oracion = "En un lugar de la Mancha"

print(len(oracion))           # numero de caracteres (incluye espacios)
print(oracion.split())        # lista de palabras
print(len(oracion.split()))   # numero de palabras
````

> 🧭 **Nota para docentes.** `len(texto.split())` es un gran ejemplo de **abstracción**:
> con una sola línea resolvemos algo que parece complejo (contar palabras). Es un "truco con
> efecto" ideal para una primera clase: el estudiante siente que logró mucho con muy poco.


## 5. Tu primer analizador de texto

Vamos a juntar todo lo de hoy en un pequeño programa que **describe** un texto. No usa
condicionales ni bucles: solo lo que ya sabes.


````python
# Texto de ejemplo (fragmento de "Don Quijote", dominio publico)
texto = "En un lugar de la Mancha de cuyo nombre no quiero acordarme"

num_caracteres = len(texto)
num_palabras = len(texto.split())
en_mayusculas = texto.upper()
veces_de = texto.lower().count("de")

print("===== ANALISIS DEL TEXTO =====")
print(f"Caracteres: {num_caracteres}")
print(f"Palabras:   {num_palabras}")
print(f"Veces que aparece 'de': {veces_de}")
print(f"En mayusculas: {en_mayusculas}")
````

Si quieres, prueba la **versión interactiva**: pide el texto a la persona usuaria.
(Puedes pegar un párrafo cualquiera cuando te lo solicite.)


````python
texto = input("Pega o escribe un texto y presiona Enter:\n")

print("\n===== ANALISIS DEL TEXTO =====")
print(f"Caracteres: {len(texto)}")
print(f"Palabras:   {len(texto.split())}")
print(f"En minusculas: {texto.lower()}")
````
