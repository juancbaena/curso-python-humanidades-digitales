### A.2 Plan B — Instalar Python y VS Code en tu computadora

Trabajaremos en Colab durante el curso, pero tener Python **local** te da independencia de la
conexión. Hazlo **con apoyo del monitor**. Pasos generales:

**1. Instalar Python**
- **Windows:** descarga desde <https://www.python.org/downloads/> y, en el instalador,
  **marca la casilla "Add Python to PATH"** antes de continuar (o instalalo desde la tienda de windows).
- **macOS:** descarga el instalador desde la misma página (o usa Homebrew: `brew install python`).

**2. Instalar Visual Studio Code**
- Descarga desde <https://code.visualstudio.com/> e instálalo.

**3. Instalar la extensión de Python en VS Code**
- Abre VS Code → icono de **Extensiones** (cuadritos en la barra lateral) → busca **"Python"**
  (de Microsoft) → **Install**.

**4. Probar que todo funciona**
- Crea un archivo `prueba.py`, escribe `print("Hola desde mi computadora")` y ejecútalo con el
