### Parte B — Extracción de la web (scraping)

**Ejecuta primero esta celda** para descargar la página de ejemplo y crear el objeto `sopa`.
Las actividades 3 a 5 lo usan.


````python
import requests
from bs4 import BeautifulSoup

html = requests.get("https://raw.githubusercontent.com/calderonf/curso-python-humanidades-digitales/main/datos/catalogo_ejemplo.html").text
sopa = BeautifulSoup(html, "html.parser")
print(sopa.title.text)
````

## Actividad 3 — Extraer títulos y autores

De la tabla con id `catalogo`, recorre las filas (`#catalogo tbody tr`) y muestra el título y el autor de cada obra (clases `.titulo` y `.autor`).

````python
# Tu solución:
````

<details>
<summary>👀 Ver solución</summary>

```python
for fila in sopa.select("#catalogo tbody tr"):
    titulo = fila.select_one(".titulo").text
    autor = fila.select_one(".autor").text
    print(f"{titulo} — {autor}")
```
</details>

## Actividad 4 — De la web a un DataFrame

Construye una lista de diccionarios con `titulo`, `autor`, `anio` (como número) y `genero` de cada fila, y conviértela en un DataFrame de pandas. Muéstralo.

````python
# Tu solución:
````

<details>
<summary>👀 Ver solución</summary>

```python
import pandas as pd
registros = []
for fila in sopa.select("#catalogo tbody tr"):
    registros.append({
        "titulo": fila.select_one(".titulo").text,
        "autor": fila.select_one(".autor").text,
        "anio": int(fila.select_one(".anio").text),
        "genero": fila.select_one(".genero").text,
    })
df = pd.DataFrame(registros)
print(df)
```
</details>
