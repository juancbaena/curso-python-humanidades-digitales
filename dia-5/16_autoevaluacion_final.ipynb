## 3) El formulario

Ejecuta la celda y responde con un número del **1 al 5** para cada enunciado.
Si escribes algo que no sea 1-5, el cuaderno te lo volverá a preguntar (así funciona como un
formulario: no avanza hasta tener una respuesta válida).

> Consejo: ten a mano los cuadernos del curso por si quieres recordar algún tema mientras respondes.


```python
# @title 3) Responder el formulario (escala 1 a 5)
def preguntar(item):
    """Pide una respuesta 1-5 y la valida, como un campo de formulario."""
    print(f"\n[{item['dia']} · {item['tema']}]")
    print(item["texto"])
    while True:
        r = input("  Tu respuesta (1-5): ").strip()
        if r in {"1", "2", "3", "4", "5"}:
            return int(r)
        print("  ⚠️  Escribe solo un número del 1 al 5.")

RESPUESTAS = {}
print("=" * 70)
print("CUESTIONARIO FINAL DE AUTOEVALUACIÓN — responde con honestidad (1 a 5)")
print("=" * 70)

for item in PREGUNTAS:
    RESPUESTAS[item["id"]] = preguntar(item)

print("\n" + "=" * 70)
print(f"✅ Has respondido los {len(RESPUESTAS)} enunciados.")
print("Continúa con la siguiente celda para ver tu diagnóstico.")
```

## 4) Tu diagnóstico y recomendaciones

Esta celda analiza tus respuestas:

- Calcula tu nivel **por día** y un nivel **global**.
- Para cada enunciado donde te calificaste **3 o menos**, te recomienda un **cuaderno concreto**
  de repaso y de contenido para reforzar.

No hay respuestas "correctas": el diagnóstico refleja tu propia percepción.


```python
# @title 4) Calcular diagnóstico
def nivel_texto(prom):
    if prom >= 4.5: return "Avanzado"
    if prom >= 3.5: return "Sólido"
    if prom >= 2.5: return "En desarrollo"
    return "Inicial — conviene reforzar"

# Promedio global y por día
todos = list(RESPUESTAS.values())
prom_global = sum(todos) / len(todos)

dias = {}
for item in PREGUNTAS:
    dias.setdefault(item["dia"], []).append(RESPUESTAS[item["id"]])
prom_dia = {d: sum(v) / len(v) for d, v in dias.items()}

# Enunciados a reforzar (calificación <= 3)
a_reforzar = [item for item in PREGUNTAS if RESPUESTAS[item["id"]] <= 3]

print("=" * 70)
print(f"DIAGNÓSTICO DE {NOMBRE.upper()}")
print("=" * 70)
print(f"\nNivel global: {prom_global:.1f}/5  ->  {nivel_texto(prom_global)}\n")

print("Nivel por día:")
for d in ["Día 1", "Día 2", "Día 3", "Día 4", "Día 5"]:
    if d in prom_dia:
        print(f"  {d}: {prom_dia[d]:.1f}/5  ->  {nivel_texto(prom_dia[d])}")

print("\n" + "-" * 70)
if a_reforzar:
    print(f"TEMAS PARA REFORZAR ({len(a_reforzar)}): revisa estos cuadernos\n")
    for item in a_reforzar:
        print(f"• {item['dia']} · {item['tema']}  (te calificaste {RESPUESTAS[item['id']]}/5)")
        print(f"    Repaso:    {item['repaso']}")
        print(f"    Contenido: {item['contenido']}")
else:
    print("¡Excelente! No marcaste ningún tema en nivel bajo.")
    print("Te recomendamos repasar el cuaderno del proyecto integrador:")
    print("    dia-5/15_proyecto_integrador_ia_local.ipynb")
print("-" * 70)
```

## 5) Generar tu PDF para enviar

Esta celda crea un archivo **PDF** con tu diagnóstico y tus recomendaciones.
Es lo que debes enviar por correo. (Si es la primera vez, instala la librería ejecutando
la primera línea; en Colab ya suele estar disponible.)


```python
# @title 5) Crear el PDF de resultados
!pip install fpdf2 -q

from fpdf import FPDF
from datetime import date

def limpia(texto):
    """Mantiene acentos del español y evita caracteres fuera de latin-1."""
    reemplazos = {"—": "-", "–": "-", "“": '"', "”": '"', "‘": "'", "’": "'",
                  "•": "-", "✅": "", "⚠️": "", "⏱️": "", "→": "->", "👀": ""}
    for a, b in reemplazos.items():
        texto = texto.replace(a, b)
    return texto.encode("latin-1", "replace").decode("latin-1")

pdf = FPDF()
pdf.add_page()
pdf.set_auto_page_break(auto=True, margin=15)

def linea(txt, h=6, estilo="", size=11):
    """Escribe una línea reseteando X al margen izquierdo (evita el error de ancho 0)."""
    pdf.set_font("Helvetica", estilo, size)
    pdf.set_x(pdf.l_margin)
    pdf.multi_cell(0, h, limpia(txt))

# Encabezado
linea("Cuestionario final de autoevaluación", h=8, estilo="B", size=15)
linea("EH1023 - Programación para Humanidades Digitales", size=10)
linea("Programar para enseñar - Javeriana / Tec de Monterrey", size=10)
pdf.ln(3)
linea(f"Nombre: {NOMBRE}")
linea(f"Programa / área: {PROGRAMA}")
linea(f"Fecha: {date.today().isoformat()}")
pdf.ln(3)

# Nivel global
linea(f"Nivel global: {prom_global:.1f}/5 - {nivel_texto(prom_global)}", h=7, estilo="B", size=12)
pdf.ln(1)

# Nivel por día
linea("Nivel por día:", estilo="B")
for d in ["Día 1", "Día 2", "Día 3", "Día 4", "Día 5"]:
    if d in prom_dia:
        linea(f"  {d}: {prom_dia[d]:.1f}/5 - {nivel_texto(prom_dia[d])}")
pdf.ln(2)

# Detalle de respuestas
linea("Detalle de tus respuestas:", estilo="B")
for item in PREGUNTAS:
    linea(f"  [{RESPUESTAS[item['id']]}/5] {item['dia']} - {item['tema']}", h=5, size=10)
pdf.ln(2)

# Recomendaciones
linea("Recomendaciones de refuerzo:", estilo="B")
if a_reforzar:
    for item in a_reforzar:
        linea(f"- {item['dia']} - {item['tema']} (te calificaste {RESPUESTAS[item['id']]}/5)", h=5, size=10)
        linea(f"    Repaso: {item['repaso']}", h=5, size=10)
        linea(f"    Contenido: {item['contenido']}", h=5, size=10)
else:
    linea("Sin temas en nivel bajo. Revisa el proyecto integrador: dia-5/15_proyecto_integrador_ia_local.ipynb", h=5, size=10)
pdf.ln(2)

linea("Nota: este documento es un diagnóstico personal de autoevaluación. No requiere "
      "enviar soluciones de los ejercicios del curso.", h=5, estilo="I", size=9)

nombre_archivo = "autoevaluacion_" + "".join(c for c in NOMBRE if c.isalnum() or c in " _-").strip().replace(" ", "_") + ".pdf"
pdf.output(nombre_archivo)
print(f"✅ PDF creado: {nombre_archivo}")
print("En Colab: panel izquierdo (icono de carpeta) -> clic derecho sobre el archivo -> Descargar.")
```
