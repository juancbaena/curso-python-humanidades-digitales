## 1. Medios digitales: manipular imágenes con Pillow

Una imagen también son **datos**: una cuadrícula de píxeles, cada uno con color. Con la librería
**Pillow** (`PIL`) podemos abrir, transformar y guardar imágenes. En Colab ya está instalada; en local:
`py -m pip install pillow`.

Seguimos el método con IA si quieres, pero aquí el código es corto. Cargamos una imagen de ejemplo del
repositorio.


```python
import urllib.request
from PIL import Image

# Descargamos la imagen de ejemplo del repositorio
urllib.request.urlretrieve("https://raw.githubusercontent.com/calderonf/curso-python-humanidades-digitales/main/datos/imagen_ejemplo.png", "imagen.png")

img = Image.open("imagen.png")
print("Tamano (ancho, alto):", img.size)
print("Modo de color:", img.mode)
img
```

### Transformaciones básicas

Tres operaciones muy usadas: **redimensionar**, pasar a **escala de grises** y **rotar**.


```python
# Redimensionar a la mitad
ancho, alto = img.size
pequena = img.resize((ancho // 2, alto // 2))
print("Nuevo tamano:", pequena.size)
pequena
```

```python
# Escala de grises
gris = img.convert("L")   # "L" = luminancia (grises)
gris
```

```python
# Rotar 90 grados y guardar el resultado
rotada = img.rotate(90, expand=True)
rotada.save("imagen_rotada.png")
print("Imagen rotada guardada como 'imagen_rotada.png'.")
rotada
```

> 🧭 **Nota para docentes.** Las imágenes conectan muy bien con estudios visuales, fotografía y cine.
> Una actividad sencilla y vistosa: tomar una carpeta de imágenes, redimensionarlas y aplicarles un
> filtro en lote. El audio y el video se manejan con la misma lógica (con librerías como `pydub` o
> `moviepy`), pero conviene dejarlos para proyectos más avanzados.
