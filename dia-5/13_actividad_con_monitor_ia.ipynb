## Menú de pistas (elige una)

### 🅰️ Análisis de texto literario o filosófico
> Para quienes marcaron *análisis de textos, voces de personajes, corpus*.
1. **Capturar:** *"Descarga el texto de esta URL y guárdalo en una variable."* (usa un .txt del repo o uno propio)
2. **Procesar:** *"Cuenta las 15 palabras más frecuentes, en minúsculas, sin signos de puntuación ni palabras vacías del español."*
3. **Visualizar:** *"Grafica en barras esas 15 palabras."*

### 🅱️ Visualización de un catálogo o playlist (CSV)
> Para quienes marcaron *visualización de datos, música/cine, datos culturales*.
1. **Capturar:** *"Carga este CSV (titulo, autor/artista, anio, genero) en un DataFrame de pandas."*
2. **Procesar:** *"Cuenta cuántos elementos hay por género y el promedio por década."*
3. **Visualizar:** *"Haz un gráfico de barras por género y uno de líneas por década."*

### 🅲️ Libros desde una API (Open Library)
> Para quienes marcaron *APIs, bases de datos, mundo editorial*.
1. **Capturar:** *"Con requests, consulta `https://openlibrary.org/search.json?author=TU_AUTOR&limit=10` y guarda el JSON."*
2. **Procesar:** *"Arma una lista de (título, año) y ordénala por año."*
3. **Visualizar:** *"Haz un gráfico de cuántos libros hay por década."*

### 🅳️ Extracción de una página web (scraping)
> Para quienes marcaron *extracción web, cobertura en prensa/redes*.
1. **Capturar:** *"Descarga el HTML de esta URL y crea un objeto BeautifulSoup."* (usa la página de ejemplo del repo o una página de la que tengas permiso de descargar datos)
2. **Procesar:** *"Extrae los títulos y autores de la tabla y arma un DataFrame."*
3. **Visualizar:** *"Cuenta las obras por género y grafícalo."*

### 🅴️ Conteo de palabras clave en titulares (texto + temas sociales)
> Para quienes marcaron *redes/medios, narrativas, problemáticas sociales*.
1. **Capturar:** *"Carga este archivo de titulares en una lista (una línea por titular)."*
2. **Procesar:** *"Cuenta en cuántos titulares aparece cada una de estas palabras clave: [escribe 3-4]."*
3. **Visualizar:** *"Haz un gráfico de barras con la frecuencia de cada palabra clave."*
