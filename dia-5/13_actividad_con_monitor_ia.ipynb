## ¿Qué vas a lograr hoy?

Al terminar tendrás, en una carpeta de **tu disco**, un programa llamado `analisis.py` que:

1. usa **pandas** (tablas de datos) y **seaborn** (gráficos bonitos),
2. dibuja un gráfico de barras, y
3. **guarda una imagen en PDF** (`grafica.pdf`) que podrás abrir e imprimir.

Y lo más importante: lo habrás **instalado y ejecutado tú**, en un **entorno virtual** propio.

> 💡 Recordatorio del Día 4: un **entorno virtual** (`venv`) es una "cajita" aislada con su propio
> Python y sus propios paquetes, solo para **este** proyecto. Así las librerías de un trabajo no
> chocan con las de otro.


## Paso 1 — Crea la carpeta del proyecto y ábrela en VS Code

1. En tu computadora, crea una carpeta nueva, por ejemplo `mi-primer-script` (en *Documentos* o en
   el *Escritorio*, donde la encuentres fácil).
2. Abre **VS Code**.
3. Menú **Archivo → Abrir carpeta…** (*File → Open Folder…*) y selecciona la carpeta que creaste.
4. Si VS Code pregunta *"¿Confías en los autores de esta carpeta?"*, responde **Sí**.

> 🧭 **Para docentes.** Trabajar **abriendo una carpeta** (y no archivos sueltos) es clave: la terminal
> y el entorno virtual quedarán "dentro" del proyecto, que es justo lo que queremos enseñar.


## Paso 2 — Crea el archivo `analisis.py`

1. En el panel izquierdo (el explorador de archivos), pulsa el icono de **Nuevo archivo**.
2. Nómbralo exactamente **`analisis.py`** (la terminación `.py` es la que lo hace un programa de Python).
3. Copia el siguiente código dentro del archivo y **guárdalo** con `Ctrl + S` (en Mac, `Cmd + S`).


```python
# analisis.py — mi primer script local con entorno virtual
# Usa tres librerias: pandas (datos), seaborn y matplotlib (graficos).

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1) Datos de ejemplo: numero de obras literarias por genero
df = pd.DataFrame({
    "genero": ["Novela", "Poesia", "Teatro", "Cuento"],
    "obras":  [5, 4, 2, 3],
})
print("Datos cargados:")
print(df)

# 2) Estilo bonito con seaborn y un grafico de barras
sns.set_theme(style="whitegrid")
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x="genero", y="obras", hue="genero",
            palette="viridis", legend=False)

plt.title("Obras literarias por genero")
plt.xlabel("Genero")
plt.ylabel("Numero de obras")
plt.tight_layout()

# 3) Guardar el grafico como PDF en la misma carpeta del script
plt.savefig("grafica.pdf")
print("Listo: se guardo 'grafica.pdf' en la carpeta del proyecto.")
```

> Si VS Code subraya `import pandas` con un foco amarillo y dice *"no se pudo resolver"*, **no te
> preocupes todavía**: aún no hemos instalado las librerías. Eso es lo que haremos en los siguientes
> pasos.


## Paso 3 — Abre la terminal y crea el entorno virtual

Abre la terminal integrada: menú **Terminal → New Terminal** (o `Ctrl + ñ`). Aparecerá abajo, ya
ubicada **dentro de tu carpeta**. Ahí escribe:

**1. Crear el entorno** (solo una vez por proyecto):
```
py -m venv .venv
```
> En **macOS / Linux** usa `python3 -m venv .venv`.

**2. Activarlo:**
- **Windows · PowerShell** (la terminal azul por defecto de VS Code):
  ```
  .\.venv\Scripts\Activate.ps1
  ```
- **Windows · Símbolo del sistema (cmd):**
  ```
  .venv\Scripts\activate.bat
  ```
- **macOS / Linux:**
  ```
  source .venv/bin/activate
  ```

Cuando el entorno esté activo verás **`(.venv)`** al inicio de la línea de la terminal. Esa es tu señal
de que estás "dentro de la cajita".

> 🛟 ¿Error *"...la ejecución de scripts está deshabilitada..."* en PowerShell? Es el **Error E** de
> la guía del Día 3: usa una terminal **cmd** y `activate.bat`, o ejecuta una vez
> `Set-ExecutionPolicy -Scope CurrentUser -ExecutionPolicy RemoteSigned`.


## Paso 4 — Instala los prerrequisitos **dentro** del entorno

Con el `(.venv)` activo, instala las tres librerías que usa el script. Aquí **sí** puedes usar `pip`
directo, porque apunta al entorno virtual:

```
pip install pandas seaborn matplotlib
```

Verás muchas líneas y, al final, `Successfully installed ...`. Puede tardar uno o dos minutos: es
normal (seaborn arrastra varias librerías).

> ✅ **Buena práctica reutilizable — `requirements.txt`.** En proyectos reales, las dependencias se
> anotan en un archivo. Crea uno llamado **`requirements.txt`** con este contenido:
>
> ```
> pandas
> seaborn
> matplotlib
> ```
>
> y luego instala todo de un golpe con:
>
> ```
> pip install -r requirements.txt
> ```
>
> Así cualquiera (¡o tus estudiantes!) puede recrear el entorno con un solo comando.


## Paso 5 — Dile a VS Code que use el entorno

Para que VS Code "vea" lo que acabas de instalar:

1. Presiona `Ctrl + Shift + P`.
2. Escribe **Python: Select Interpreter** y selecciónalo.
3. Elige el que diga **`.venv`** (suele aparecer como *recomendado*).

Si ahora el subrayado amarillo de los `import` desaparece, vas perfecto.


## Paso 6 — Ejecuta el script

Tienes dos formas, elige la que prefieras:

- **Con el botón ▶** (*Run Python File*) arriba a la derecha del editor, o
- **Desde la terminal** (con el `(.venv)` activo):
  ```
  python analisis.py
  ```
  > En Windows, si `python` no responde, prueba `py analisis.py`.

Si todo salió bien, en la terminal verás la tabla de datos y el mensaje
**`Listo: se guardo 'grafica.pdf' en la carpeta del proyecto.`**


## Paso 7 — Abre tu PDF

En el explorador de archivos de VS Code (panel izquierdo) verás un archivo nuevo: **`grafica.pdf`**.
Haz doble clic para abrirlo (o ábrelo desde la carpeta en tu computadora). Deberías ver tu gráfico de
barras de **obras por género**.

🎉 **¡Lo lograste!** Creaste una carpeta, escribiste un programa, montaste un entorno virtual,
instalaste librerías y generaste un PDF — todo en tu propia máquina.


## (Opcional) Vista previa aquí mismo

Si quieres **ver** el resultado sin salir de este cuaderno, ejecuta la celda de abajo: hace lo mismo
que tu script y muestra el gráfico en pantalla. *(En tu computadora local, el código vive en
`analisis.py`, no aquí.)*


````python
# Vista previa del mismo analisis (equivale a 'analisis.py')
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "genero": ["Novela", "Poesia", "Teatro", "Cuento"],
    "obras":  [5, 4, 2, 3],
})

sns.set_theme(style="whitegrid")
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x="genero", y="obras", hue="genero",
            palette="viridis", legend=False)
plt.title("Obras literarias por genero")
plt.xlabel("Genero")
plt.ylabel("Numero de obras")
plt.tight_layout()
plt.savefig("grafica.pdf")    # se guarda junto a este cuaderno
plt.show()
print("Se guardo 'grafica.pdf'.")
````
