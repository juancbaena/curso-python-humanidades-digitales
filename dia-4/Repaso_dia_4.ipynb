**Ejecuta primero esta celda** para capturar los datos. Las actividades 3 a 5 usan `df`.


````python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/calderonf/curso-python-humanidades-digitales/main/datos/obras_literarias.csv")
df.head()
````

## Actividad 3 — Procesar: promedio por década

Con `df`, crea una columna `decada` (el año redondeado a la decena inferior: `(anio // 10) * 10`) y cuenta cuántas obras hay por década.

````python
# Tu solución:
````

<details>
<summary>👀 Ver solución</summary>

```python

print(df["decada"].value_counts().sort_index())
```
</details>

## Actividad 4 — Visualizar: obras por década

Haz un gráfico de **líneas** del número de obras por década (usa la columna `decada` de la actividad anterior).

````python
# Tu solución:
````

<details>
<summary>👀 Ver solución</summary>

```python
import matplotlib.pyplot as plt
por_decada = df["decada"].value_counts().sort_index()
plt.figure(figsize=(9, 4))
plt.plot(por_decada.index, por_decada.values, marker="o", color="darkgreen")
plt.title("Obras por decada")
plt.xlabel("Decada")
plt.ylabel("Numero de obras")
plt.tight_layout()
plt.show()
```
</details>

## Actividad 5 — Pipeline de texto: palabras frecuentes

Descarga el fragmento literario del repositorio (te damos el código), cuéntalo en minúsculas sin puntuación y muestra las **8 palabras más frecuentes**.

````python
import urllib.request
from collections import Counter

texto = urllib.request.urlopen("https://raw.githubusercontent.com/calderonf/curso-python-humanidades-digitales/main/datos/fragmento_literario.txt").read().decode("utf-8")

# Tu solución:
````

<details>
<summary>👀 Ver solución</summary>

```python
limpio = texto.lower().replace(",", "").replace(".", "")
palabras = limpio.split()
print(Counter(palabras).most_common(8))
```
</details>
