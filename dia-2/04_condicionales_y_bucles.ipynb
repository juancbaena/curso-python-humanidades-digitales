## 1. Comparar valores: los booleanos

Antes de decidir, hay que **comparar**. Una comparación devuelve un **booleano**: `True` (verdadero)
o `False` (falso). Operadores de comparación:

| Operador | Significado |
|---|---|
| `==` | ¿son iguales? |
| `!=` | ¿son distintos? |
| `<`  `>` | menor que / mayor que |
| `<=` `>=` | menor o igual / mayor o igual |

> ⚠️ Un solo `=` **asigna** una variable; el doble `==` **compara**. Confundirlos es uno de los
> errores más comunes al empezar.


````python
anio = 1605

print(anio == 1605)   # ¿es igual a 1605?
print(anio > 1500)    # ¿es mayor que 1500?
print(anio < 1000)    # ¿es menor que 1000?
print("Quijote" != "Quixote")  # ¿son distintos?
````

**Otro ejemplo (comparar cadenas de texto).** Las comparaciones también funcionan con cadenas:
`==` distingue mayúsculas de minúsculas, y `<` / `>` comparan en **orden alfabético**.

````python
autor1 = "Borges"
autor2 = "borges"

print(autor1 == autor2)                 # False: la mayuscula importa
print(autor1.lower() == autor2.lower())  # True: iguales si ignoramos mayusculas
print("Alfaro" < "Borges")              # True: la A va antes que la B
print("Cervantes" > "Borges")           # True: la C va despues que la B
````

## 2. Tomar decisiones: `if`, `elif`, `else`

Un **condicional** ejecuta un bloque de código **solo si** una condición es verdadera.
Fíjate en dos detalles que en Python son obligatorios:

1. La línea del `if` termina en **dos puntos** `:`
2. El código que depende del `if` va **indentado** (con sangría, normalmente 4 espacios).

La sangría es la forma en que Python sabe qué instrucciones están "dentro" de la decisión.


````python
palabra = "humanidades"

if len(palabra) <= 3:
    categoria = "corta"
elif len(palabra) <= 7:
    categoria = "media"
else:
    categoria = "larga"

print(f"La palabra '{palabra}' es {categoria} ({len(palabra)} letras).")
````

> 🧭 **Nota para docentes.** La **indentación** es la primera gran fuente de errores y de
> frustración para quien empieza. Conviene mostrarlo a propósito: un espacio de más o de menos
> cambia el significado del programa. Anticipar este tropiezo —en vez de dejar que el estudiante
> choque solo con él— es justo el tipo de cuidado que evita que se desanime.


Un ejemplo más humanístico: clasificar un **año** en un período histórico. Aquí se ve la
potencia del `elif` para encadenar varias decisiones.


````python
anio = 1492

if anio < 476:
    periodo = "Edad Antigua"
elif anio < 1492:
    periodo = "Edad Media"
elif anio < 1789:
    periodo = "Edad Moderna"
else:
    periodo = "Edad Contemporanea"

print(f"El anio {anio} corresponde a la {periodo}.")
````

**Otro ejemplo (decidir comparando texto, con un caso realista).** Hasta aquí decidimos con números
(longitud, año). También podemos decidir **comparando cadenas**. Pero en datos reales el texto llega
"sucio": el género puede venir como `"Poesía"`, `"POESÍA"` o `"poesia"`. Para que la comparación
funcione en todos los casos, primero **normalizamos**: pasamos todo a minúsculas y **quitamos las
tildes** encadenando varios `.replace(...)`. Recién entonces comparamos.

````python
# Un caso real: el genero puede venir con MAYUSCULAS y con tildes ("Poesía", "POESÍA"...).
# Antes de comparar, lo "normalizamos": todo a minusculas y sin tildes.
genero = "Poesía"

genero_normalizado = (genero.lower()
                            .replace("á", "a")
                            .replace("é", "e")
                            .replace("í", "i")
                            .replace("ó", "o")
                            .replace("ú", "u"))

if genero_normalizado == "novela":
    estante = "Narrativa"
elif genero_normalizado == "poesia":
    estante = "Lírica"
elif genero_normalizado == "teatro":
    estante = "Dramaturgia"
else:
    estante = "General"

# Mostramos el genero original, pero la decision se tomo con la version normalizada.
print(f"La obra de genero '{genero}' va en el estante: {estante}")

# Prueba cambiando 'genero' por "POESÍA", "Novela" o "teatro": funciona en todos los casos.
````

## 3. Repetir con `for`: recorrer una colección

Un bucle `for` ejecuta el mismo bloque **una vez por cada elemento** de una colección.
Una **lista** es una colección de valores entre corchetes `[ ]` (mañana las veremos a fondo;
hoy solo la usamos para recorrerla).


````python
autores = ["Cervantes", "Sor Juana", "Borges", "Rulfo"]

for autor in autores:
    print(f"Autor en estudio: {autor}")
````

También se puede recorrer **una cadena**, carácter por carácter:


````python
for letra in "Mancha":
    print(letra)
````

### Repetir un número fijo de veces: `range()`

`range(n)` genera los números de `0` a `n-1`. Es ideal para repetir algo `n` veces.


````python
for numero in range(5):
    print(f"Verso número {numero + 1}")
````

**Otro ejemplo (recorrer números y acumular).** Un `for` también sirve para **sumar** los valores
de una lista de números, distinto a recorrer textos.

````python
minutos = [3, 5, 2, 8]   # minutos de lectura de cuatro poemas

total = 0
for m in minutos:
    total = total + m

print(f"Tiempo total de lectura: {total} minutos")
````

## 4. Repetir con `while`: hasta que se cumpla una condición

El bucle `while` repite **mientras** una condición sea verdadera. Hay que asegurarse de que la
condición termine cambiando, o el bucle no se detendría nunca.


````python
contador = 1

while contador <= 3:
    print(f"Lectura en voz alta, ronda {contador}")
    contador = contador + 1   # este cambio evita que el bucle sea infinito
print("Fin de la actividad.")
````

**Otro ejemplo (repetir hasta vaciar una lista).** Aquí el `while` se repite **mientras queden
tareas**: el método `.pop(0)` saca la primera de la lista hasta dejarla vacía.

````python
pendientes = ["leer", "anotar", "comentar"]

while len(pendientes) > 0:
    tarea = pendientes.pop(0)   # saca y devuelve la primera tarea
    print("Haciendo:", tarea)

print("No quedan tareas pendientes.")
````

## 5. La combinación poderosa: `for` + `if`

Aquí está el corazón de la automatización: **recorrer** muchos elementos y **decidir** algo en
cada uno. Vamos a clasificar una lista de años por período histórico.


````python
anios = [1325, 1492, 1605, 1821, 1910]

for anio in anios:
    if anio < 476:
        periodo = "Edad Antigua"
    elif anio < 1492:
        periodo = "Edad Media"
    elif anio < 1789:
        periodo = "Edad Moderna"
    else:
        periodo = "Edad Contemporanea"
    print(f"{anio}: {periodo}")
````

**Otro ejemplo (`for` + `if` con texto).** Antes contamos por año (número); ahora contamos títulos
que **empiezan con vocal**, una condición sobre cadenas.

````python
titulos = ["Aura", "El tunel", "Insomnio", "Pedro Paramo", "Otono"]

empiezan_vocal = 0
for t in titulos:
    if t[0].lower() in "aeiou":
        empiezan_vocal = empiezan_vocal + 1

print(f"Titulos que empiezan con vocal: {empiezan_vocal}")
````

## 6. Trabajar con datos reales del repositorio

De aquí en adelante, los archivos de datos viven en la carpeta **`datos/`** del repositorio del
curso. **No necesitas descargar nada a mano**: la siguiente línea trae el archivo directamente a
tu entorno desde internet.

Usamos la dirección **"raw"** (texto plano) del archivo. El patrón es:

```
https://raw.githubusercontent.com/calderonf/curso-python-humanidades-digitales/main/datos/NOMBRE_DEL_ARCHIVO
```

> 💡 No te preocupes por entender estas dos líneas todavía: por ahora solo **ejecútalas**.
> Sirven para "traer" el texto y guardarlo en la variable `texto`.


````python
import urllib.request

url = "https://raw.githubusercontent.com/calderonf/curso-python-humanidades-digitales/main/datos/fragmento_literario.txt"
texto = urllib.request.urlopen(url).read().decode("utf-8")

print(texto)
````

Ahora que el texto está en tu entorno, combinemos todo lo de hoy: lo separamos en palabras con
`.split()` y **recorremos** las palabras contando cuántas son **largas** (más de 6 letras).


````python
palabras = texto.split()

total = 0
palabras_largas = 0

for palabra in palabras:
    total = total + 1
    if len(palabra) > 6:
        palabras_largas = palabras_largas + 1

print(f"Palabras en total: {total}")
print(f"Palabras largas (mas de 6 letras): {palabras_largas}")
````

**Otro ejemplo (buscar un máximo).** Sobre el mismo texto, en vez de contar, **buscamos** la palabra
más larga recorriéndolas y quedándonos con la mayor encontrada.

````python
palabras = texto.split()

mas_larga = ""
for palabra in palabras:
    if len(palabra) > len(mas_larga):
        mas_larga = palabra

print("Palabra mas larga del fragmento:", mas_larga)
````
