### A.2 — Recorrer una lista con `for`

Dada la lista `obras = ["La Iliada", "El Quijote", "Rayuela"]`, recórrela e imprime cada obra
precedida por `"Leemos:"`.


````python
obras = ["La Iliada", "El Quijote", "Rayuela"]

# Tu solucion A.2:
````

### A.3 — `for` + `if` (contar)

Dada la lista `anios = [1320, 1605, 1870, 1920, 2001]`, cuenta cuántos años son **posteriores a 1900**
y muestra el total.


````python
anios = [1320, 1605, 1870, 1920, 2001]

# Tu solucion A.3:
````

<details>
<summary>👀 Solución posible A.3</summary>

```python
anios = [1320, 1605, 1870, 1920, 2001]
posteriores = 0
for anio in anios:
    if anio > 1900:
        posteriores = posteriores + 1
print(f"Anios posteriores a 1900: {posteriores}")
```
</details>


### A.4 — (Opcional) Practicar con datos del repositorio

Si te quedó tiempo, trae el archivo de fechas del repositorio y recórrelo. Por ahora solo
**ejecuta** la primera celda (trae los datos) y luego intenta el conteo.


````python
import urllib.request

url = "https://raw.githubusercontent.com/calderonf/curso-python-humanidades-digitales/main/datos/fechas_historicas.csv"
contenido = urllib.request.urlopen(url).read().decode("utf-8")
print(contenido)
````

Cada línea tiene la forma `evento,anio`. Con `.splitlines()` separamos por líneas. **Reto guiado:**
recorre las líneas (saltando la primera, que es el encabezado) e imprime solo el nombre del evento.


````python
lineas = contenido.splitlines()

# Tu intento (pista: lineas[1:] omite el encabezado; linea.split(",")[0] es el evento):
````

<details>
<summary>👀 Solución posible A.4</summary>

```python
lineas = contenido.splitlines()
for linea in lineas[1:]:        # saltamos el encabezado
    evento = linea.split(",")[0]
    print(evento)
```
