## 1. Funciones: empaquetar código para reutilizarlo

Una **función** es un bloque de código con nombre que realiza una tarea. La defines una vez y la
**llamas** las veces que quieras. Se crea con `def`, recibe **parámetros** entre paréntesis y suele
**devolver** un resultado con `return`.

Piensa en una función como una **receta**: le das ingredientes (parámetros) y te entrega un plato (`return`).


````python
def saludar(nombre):
    mensaje = f"Hola, {nombre}. ¡Bienvenido al curso!"
    return mensaje

# Llamamos a la funcion (la usamos):
print(saludar("Ana"))
print(saludar("Luis"))
````

Una función nos deja **reutilizar** una lógica. Recuerda la clasificación de palabras de ayer:
ahora la encerramos en una función y la usamos cuantas veces queramos.


````python
def clasificar_palabra(palabra):
    if len(palabra) <= 3:
        return "corta"
    elif len(palabra) <= 7:
        return "media"
    else:
        return "larga"

print(clasificar_palabra("sol"))
print(clasificar_palabra("cultura"))
print(clasificar_palabra("humanidades"))
````

**Otro ejemplo (una función con dos parámetros que calcula).** Una función puede recibir varios
datos y devolver un número, como la antigüedad de una obra.

````python
def antiguedad(anio_publicacion, anio_actual):
    return anio_actual - anio_publicacion

print(antiguedad(1605, 2026), "anios")
print(antiguedad(1967, 2026), "anios")
````

## 2. `return` no es lo mismo que `print`

Este punto confunde a casi todo el mundo al empezar:

- `print` **muestra** algo en pantalla, pero no lo guarda para seguir trabajando.
- `return` **devuelve** un valor que puedes **guardar en una variable** y reutilizar.


````python
def doble_con_print(n):
    print(n * 2)        # solo lo muestra

def doble_con_return(n):
    return n * 2        # lo devuelve para poder usarlo

doble_con_print(5)              # muestra 10, pero no lo puedo reutilizar
resultado = doble_con_return(5) # guardo el 10 en una variable
print(resultado + 1)            # y lo uso: 11
````

**Otro ejemplo (reutilizar lo que se devuelve, con texto).** Como `limpiar` **devuelve** un valor,
podemos guardarlo y seguir trabajando con él.

````python
def limpiar(texto):
    return texto.strip().lower()

titulo = limpiar("   EL Quijote   ")
print(titulo)
print("Empieza con 'el':", titulo.startswith("el"))   # reusamos el valor devuelto
````

> 🧭 **Nota para docentes.** La diferencia entre `return` y `print` es un clásico tropiezo. Una
> imagen que funciona en clase: `print` es *decir* el resultado en voz alta; `return` es *entregarlo*
> en la mano para seguir usándolo. Vale la pena dedicarle un minuto: ahorra muchísima confusión después.


## 3. Listas: una colección ordenada

Una **lista** guarda varios valores en orden, entre corchetes o parentesis cuadrado  `[ ]`. Ya las recorrimos ayer; hoy
aprendemos a **modificarlas**.


````python
autores = ["Cervantes", "Sor Juana", "Borges"]

print(autores[0])          # primer elemento
print(len(autores))        # cuantos hay

autores.append("Rulfo")    # agregar al final
print(autores)

autores.sort()             # ordenar alfabeticamente (si estas trabajando en colab usa las funciones de IA para preguntar por qué reglas sigue este ordenamiento y que es ordenamiento "in place")
print(autores)
````

**Otro ejemplo (una lista de números).** Las listas no solo guardan texto. Con una lista de números
podemos sumar, hallar el máximo o tomar una porción (*slicing*).

````python
paginas = [120, 471, 64, 736, 158]

print("Total de paginas:", sum(paginas))
print("La obra mas extensa tiene:", max(paginas), "paginas")
print("Las dos primeras:", paginas[:2])
````

## 4. Diccionarios: pares de clave y valor

Un **diccionario** asocia una **clave** con un **valor**, entre llaves o Tirantes `{ }`. Es perfecto para un
**glosario**: la clave, llave o "key" es el término y el valor su definición.


````python
glosario = {
    "algoritmo": "secuencia de pasos para resolver un problema",
    "variable": "un nombre que guarda un valor",
}

# Acceder por su clave:
print(glosario["algoritmo"])

# Agregar una entrada nueva:
glosario["bucle"] = "estructura que repite instrucciones"

# Recorrer todo el glosario:
print("Contenido Del Glosario:")
for termino in glosario:
    print(f"{termino}: {glosario[termino]}")
````

**Otro ejemplo (un diccionario con valores numéricos).** Las claves y los valores pueden ser de
cualquier tipo: aquí asociamos cada obra con su año de publicación.

````python
publicacion = {
    "Don Quijote": 1605,
    "Pedro Paramo": 1955,
    "Rayuela": 1963,
}

print("Pedro Paramo se publico en", publicacion["Pedro Paramo"])

for obra in publicacion:
    print(f"{obra}: {publicacion[obra]}")
````

## 5. Tuplas y conjuntos (los otros dos)

- Una **tupla** es como una lista pero **no se puede modificar** (es *inmutable*). Sirve para datos
  que no deberían cambiar, como un par fijo.
- Un **conjunto** (`set`) guarda elementos **sin repetir**. Es ideal para obtener los **valores únicos**.


````python
# Tupla: autor y año que van juntos y no cambian
obra = ("Don Quijote", 1605)
print(obra[0], "se publico en", obra[1])

# Conjunto: quita los repetidos automaticamente
palabras = ["el", "mar", "el", "cielo", "el", "mar"]
unicas = set(palabras)
print(unicas)
print(f"Palabras distintas: {len(unicas)}")
````

**Otro ejemplo (operaciones entre conjuntos).** Una utilidad muy potente de los conjuntos es
**comparar** dos colecciones: qué tienen en común y qué hay en total.

````python
poema_a = {"mar", "cielo", "viento", "sal"}
poema_b = {"cielo", "tierra", "viento"}

print("Palabras en comun:", poema_a & poema_b)   # interseccion
print("Todas las palabras:", poema_a | poema_b)  # union
````

## 6. Todo junto: un analizador de texto reutilizable

Vamos a escribir **una función** que reciba un texto y devuelva un pequeño **reporte** (un diccionario)
con su número de palabras y cuántas son distintas. Combina funciones + listas + conjuntos + diccionarios.


````python
def analizar_texto(texto):
    palabras = texto.lower().split()
    reporte = {
        "total_palabras": len(palabras),
        "palabras_distintas": len(set(palabras)),
    }
    return reporte

# Probamos con una frase corta:
print(analizar_texto("el mar y el cielo y el viento"))
````

**Otro ejemplo (una función que resume números).** El mismo patrón —devolver un diccionario como
reporte— sirve para datos numéricos, como una lista de cantidades de páginas.

````python
def resumen_numeros(numeros):
    return {
        "cuantos": len(numeros),
        "total": sum(numeros),
        "promedio": round(sum(numeros) / len(numeros), 1),
    }

print(resumen_numeros([120, 471, 64, 736]))
````

Ahora lo aplicamos a un **texto real** del repositorio. Solo **ejecuta** la primera celda
(trae los datos) y luego usa tu función.


````python
import urllib.request

url = "https://raw.githubusercontent.com/calderonf/curso-python-humanidades-digitales/main/datos/fragmento_literario.txt"
texto = urllib.request.urlopen(url).read().decode("utf-8")

reporte = analizar_texto(texto)
print("Reporte del fragmento:")
print(f"  Palabras en total: {reporte['total_palabras']}")
print(f"  Palabras distintas: {reporte['palabras_distintas']}")
````
