## Actividad 9 — Función con return

Escribe una función `clasificar_palabra(palabra)` que devuelva `"corta"` si tiene 3 letras o menos, `"media"` si tiene hasta 7, y `"larga"` en otro caso. Pruébala con tres palabras.

````python
# Tu solución:
````

<details>

    if len(palabra) <= 3:
        return "corta"

    else:
        return "larga"

## Actividad 13 — Reto integrador — con datos del repositorio

Descarga el archivo `fechas_historicas.csv` del repositorio (te damos el código). Recórrelo saltando el encabezado y, usando un `conjunto`, cuenta cuántos **siglos distintos** aparecen (siglo = (anio - 1) // 100 + 1). Muestra también el conjunto de siglos.

````python
import urllib.request

url = "https://raw.githubusercontent.com/calderonf/curso-python-humanidades-digitales/main/datos/fechas_historicas.csv"
texto = urllib.request.urlopen(url).read().decode("utf-8")
lineas = texto.splitlines()

# Tu solución (usa lineas[1:] para saltar el encabezado):
````

<details>
<summary>👀 Ver solución</summary>

```python
siglos = set()
for linea in lineas[1:]:
    partes = linea.split(",")
    anio = int(partes[1])
    siglo = (anio - 1) // 100 + 1
    siglos.add(siglo)

print("Siglos distintos:", len(siglos))
print("Conjunto de siglos:", siglos)
```
</details>
